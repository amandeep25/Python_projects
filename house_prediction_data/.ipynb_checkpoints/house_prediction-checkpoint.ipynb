```python
#handling Missing values categorical varaibles have dtype as object

features_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes == 'O']

for feature in features_nan:
    print(" {}: {}% missing values".format(feature, np.round(dataset[feature].isnull().mean(),4)))
```

```python
def replace(dataset,features_nan):
    data = dataset.copy()
    data[features_nan] = data[features_nan].fillna('Missing')
    return data

dataset = replace(dataset,features_nan)
dataset[features_nan].isnull().sum()
```

```python
dataset.head(5)
```

```python
#handling Missing values numerical varaibles have dtype as not object

numerical_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes != 'O']

for feature in numerical_nan:
    print(" {}: {}% missing values".format(feature, np.round(dataset[feature].isnull().mean(),4)))
```

```python
# replace numerical missing values

for feature in numerical_nan:
    median_value = dataset[feature].median()
    
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)

dataset[numerical_nan].isnull().sum()
    
```

```python
pd.pandas.set_option('display.max_columns', None)
```

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature] = dataset['YrSold']-dataset[feature]
```

```python
dataset.head(5)
```

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

```python
num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']

for feature in num_features:
    dataset[feature] = np.log(dataset[feature])
```

```python
dataset.head()
```

```python
#handling rare categorical values say less than 1% observations
```

```python
for feature in categorical_features:
    temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df = temp[temp>0.01].index
    dataset[feature] = np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(50)
```

```python
for feature in categorical_features:
    labels_ordered = dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered = {k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature] = dataset[feature].map(labels_ordered)
```

```python
dataset.head(50)
```

```python
scaling_feature = [feature for feature in dataset.columns if feature not in ['Id','SalePrice']]
print(scaling_feature)
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
y_train = dataset['SalePrice']
```

```python
X_train = dataset.drop(['Id','SalePrice'],axis = 1)
```

```python
feature_select = SelectFromModel(Lasso(alpha = 0.005,random_state=0))
feature_select.fit(X_train,y_train)
```

```python
feature_select.get_support()
```

```python
# this is how we can make a list of the selected features
selected_feat = X_train.columns[(feature_select.get_support())]

# let's print some stats
print('total features: {}'.format((X_train.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
# print('features with coefficients shrank to zero: {}'.format(
#     np.sum(estimator_.coef_ == 0)))
```
